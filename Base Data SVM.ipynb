```python
X = df[['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal']].values.tolist()
#y = df['num'].map({0:0,1:1,2:1,3:1,4:1})
y = df['num']

import sklearn.preprocessing

#X = sklearn.preprocessing.scale(X)

def get_cross_val_scores(classifier,x,y):
    print str(classifier)[:str(classifier).find('(')]
    print 'Accuracy: %s' % str(cross_val_score(classifier,x,y,scoring='accuracy').mean())
    print 'Precision: %s' % str(cross_val_score(classifier,x,y,scoring='precision').mean())
    print 'Recall: %s' % str(cross_val_score(classifier,x,y,scoring='recall').mean())
    print 'f1: %s' % str(cross_val_score(classifier,x,y,scoring='f1').mean())
    print '\n'

   
get_cross_val_scores(LogisticRegression(),X,y)
get_cross_val_scores(KNeighborsClassifier(n_neighbors=15),X,y)
get_cross_val_scores(GaussianNB(),X,y)
get_cross_val_scores(SVC(),X,y)
get_cross_val_scores(DecisionTreeClassifier(),X,y)
get_cross_val_scores(RandomForestClassifier(),X,y)
```
